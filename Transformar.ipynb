# Definición de la función homologar

```python
def homologar(lista, min_valor, data_Frame, columna):
  """
  Description:
    La función trata de reemplazar las celdas con valores mal escritos 
    por el valor correcto
    
  Parameters
  ----------
  lista : (list) 
    Lista de strings con el formato correcto 
  
  min_valor: (int)
    Valor mínimo para buscar coincidencia en las cadenas
  
  data_Frame: (DataFrame)
    DataFrame para corregir los datos
    
  columna: (string)
    Columna a corregir del DataFrame 
  
  Returns
  -------
  """
  #min_valor = 91

  for valor in lista:
    coincidencias_Potenciales = process.extract(
      valor,
      data_Frame[columna],
      limit=data_Frame.shape[0]
    )

    for coincidencia in coincidencias_Potenciales:
      if coincidencia[1] >= min_valor:
        data_Frame.loc[data_Frame[columna]==coincidencia[0],columna] = valor
```
